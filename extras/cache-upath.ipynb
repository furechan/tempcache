```python
import os
from upath import UPath
from tempcache import TempCache
```

```python

cache = TempCache("tempcache")

# Clear all cached items for a fresh start
cache.clear_items(True)

!ls {cache.path}
```

```python
TEST_BUCKET=os.getenv("TEMPCACHE_TEST_BUCKET")

assert TEST_BUCKET is not None

TESTROOT=f"s3://{TEST_BUCKET}/tempcache"
testroot = UPath(TESTROOT)
```

```python
SAMPLE="""Some sample data for testing..."""

testroot.joinpath("sample.txt").write_text(SAMPLE)

list(i.name for i in testroot.iterdir())
```

```python
sample_path = testroot / "sample.txt"
cache_path = cache.cache_upath(sample_path)
cache_path
```
